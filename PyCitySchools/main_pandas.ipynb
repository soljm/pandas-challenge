## Setup

```python
# setup
from pathlib import Path
import pandas as pd
```

```python
# set path
school_file = Path("../Resources/schools_complete.csv")
students_file = Path("../Resources/students_complete.csv")

# read and store in DataFrames (DF)
school_df = pd.read_csv(school_file, encoding='utf-8').set_index(["school_name"])
students_df = pd.read_csv(students_file, encoding='utf-8')

# merge into one DF
complete_df = pd.merge(school_df, students_df, how="left", on=["school_name", "school_name"])
complete_df
```

## Local Government Area (LGA) Summary

```python
# calculate totals 
# total unique schools
school_count = complete_df['school_name'].nunique()
print("Total unique schools: " + str(school_count))

# total students
student_count = complete_df["student_name"].count()
print("Total students: " + str(student_count))

# total budget
total_budget = school_df["budget"].sum()
print("Total budget: " + str(total_budget))
```

```python
# calculate averages
# average maths score
avg_maths = complete_df["maths_score"].mean()
avg_maths = avg_maths.round(decimals = 2)
print("Average maths score: " + str(avg_maths))

#average reading score
avg_reading = complete_df["reading_score"].mean()
avg_reading = avg_reading.round(decimals = 2)
print("Average reading score: " + str(avg_reading))
```

```python
# calculate % passing 
# % passing maths
passing_maths = complete_df[complete_df["maths_score"] >= 50]["student_name"].count()
passing_maths_percent = (passing_maths / float(student_count)) * 100
print("Percentage of students passing maths: " + str(passing_maths_percent) + "%")

# % passing reading
passing_reading = complete_df[complete_df["reading_score"] >= 50]["student_name"].count()
passing_reading_percent = (passing_reading / float(student_count)) * 100
print("Percentage of students passing reading: " + str(passing_reading_percent) + "%")

# % overall passing
passing_overall = complete_df[(complete_df["reading_score"] >= 50) & (complete_df["maths_score"] >= 50)]["student_name"].count()
passing_overall_percent = (passing_overall / float(student_count)) * 100
print("Percentage of students passing reading: " + str(passing_overall_percent) + "%")
```

```python
# convert to DF
area_summary = pd.DataFrame({"Total Unique Schools": school_count,
                             "Total Students": student_count,
                             "Total Budget": total_budget,
                             "Average Maths Score": avg_maths,
                             "Average Reading Score": avg_reading,
                             "% Passing Maths": passing_maths_percent,
                             "% Passing Reading": passing_reading_percent,
                             "% Overall Passing": passing_overall_percent}, index=[0])
# formatting
area_summary["Total Students"] = area_summary["Total Students"].map("{:,}".format)
area_summary["Total Budget"] = area_summary["Total Budget"].map("${:,.2f}".format)
area_summary["% Passing Maths"] = area_summary["% Passing Maths"].map("{:,.2f}%".format)
area_summary["% Passing Reading"] = area_summary["% Passing Reading"].map("{:,.2f}%".format)
area_summary["% Overall Passing"] = area_summary["% Overall Passing"].map("{:,.2f}%".format)

# print
area_summary
```

## School Summary

```python
# use code provided to select type of school from school_df
school_types = school_df["type"]
school_types
```

```python
# calculate totals
# total students
students_per_school = school_df["size"]
students_per_school
```

```python
# total school budget
total_school_budget = school_df["budget"]
total_school_budget
```

```python
# budget per student
budget_per_student = total_school_budget / students_per_school
budget_per_student
```

```python
# groupby school name
school_grouped_df = complete_df.groupby(["school_name"])

# average maths score 
per_avg_math = school_grouped_df["maths_score"].mean()
per_avg_math = per_avg_math.round(decimals=2)
per_avg_math
```

```python
# average reading score 
per_avg_reading = school_grouped_df["reading_score"].mean()
per_avg_reading = per_avg_reading.round(decimals=2)
per_avg_reading
```

```python
# total students passing maths
per_pass_maths = complete_df[complete_df["maths_score"] >= 50]
per_pass_maths = per_pass_maths.groupby(["school_name"])
per_pass_maths = per_pass_maths["student_name"].count()
per_pass_maths
```

```python
# % passing maths
per_percent_maths = (per_pass_maths / students_per_school) * 100
per_percent_maths
```

```python
# total students passing reading
per_pass_reading = complete_df[complete_df["reading_score"] >= 50]
per_pass_reading = per_pass_reading.groupby(["school_name"])
per_pass_reading = per_pass_reading["student_name"].count()
per_pass_reading
```

```python
# % passing reading
per_percent_reading = (per_pass_reading / students_per_school) * 100
per_percent_reading
```

```python
# total students passing maths and reading
per_pass_overall = complete_df[(complete_df["reading_score"] >= 50) & (complete_df["maths_score"] >= 50)]
per_pass_overall = per_pass_overall.groupby(["school_name"])
per_pass_overall = per_pass_overall["student_name"].count()
per_pass_overall
```

```python
# % passing reading and maths
per_percent_overall = (per_pass_overall / students_per_school) * 100
per_percent_overall
```

```python
# create DF
school_summary = pd.DataFrame({"School Type": school_types,
                             "Total Students": students_per_school,
                             "Total School Budget": total_school_budget,
                             "Budget Per Student": budget_per_student,
                             "Average Maths Score": per_avg_math,
                             "Average Reading Score": per_avg_reading,
                             "% Passing Maths": per_percent_maths,
                             "% Passing Reading": per_percent_reading,
                             "% Overall Passing": per_percent_overall})

# formatting
school_summary["Total School Budget"] = school_summary["Total School Budget"].map("${:,.2f}".format)
school_summary["Budget Per Student"] = school_summary["Budget Per Student"].map("${:,.2f}".format)
school_summary["% Passing Maths"] = school_summary["% Passing Maths"].map("{:,.2f}%".format)
school_summary["% Passing Reading"] = school_summary["% Passing Reading"].map("{:,.2f}%".format)
school_summary["% Overall Passing"] = school_summary["% Overall Passing"].map("{:,.2f}%".format)

school_summary
```
